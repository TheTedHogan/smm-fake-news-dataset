## Let's Merge

### 1. User Data -> Followers and Followee for Each user

#### 1.1 Followers for a user

```python
followers_df= userUser.groupby('Followee')['Follower'].count().reset_index()
followers_df.columns= ['userID', 'No. of Followers']
followers_df.head()
```

```python
followers_df.shape
```

#### 1.2 Followees for a user

```python
followees_df= userUser.groupby('Follower')['Followee'].count().reset_index()
followees_df.columns= ['userID', 'No. of Followees']
followees_df.head()
```

```python
user
```

#### 1.3 Merge with followers_df

```python
users_info_0= user.merge(followers_df, how='left', on ='userID')
users_info_0
```

#### 1.4 Merge with followees_df

```python
users_info = users_info_0.merge(followees_df, how='left', on ='userID')
users_info['No. of Followers'].fillna(0, inplace = True)
users_info['No. of Followees'].fillna(0, inplace = True)
users_info.drop(columns= ['Unnamed: 0', 'userName'], inplace = True)
users_info
```

## 2. Merge User info & news info

### 2.1 Merge with newsID and postIterations

```python
userNewsInfo_0= users_info.merge(newsUserGroupped, how='right', on ='userID')
userNewsInfo_0
```

### 2.2 Merge with new Content 

```python
userNewsInfo= userNewsInfo_0.merge(newsWithContent, how='left', on ='newsID')
newsWithContent.info()
```

```python
groupped= userNewsInfo.groupby(['newsID']).agg({'No. of Followers':'mean', 'No. of Followees':'mean','PostIterations': 'sum', 'userID':'count'}).reset_index()
groupped
```
